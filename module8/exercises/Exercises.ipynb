There is heavy correlations in this dataset as well as a lot of varaibles. So, I think it would be appropriate to use PCA to reduce number of dimensions

```r
pca <- prcomp(bc[,3:ncol(bc)], center = TRUE, scale = TRUE)
plot(pca, type ="lines")
```

PCA takes features of a dataset and transforms them into 'Principal Components', capturing the variance within the data set in less componenets that originally presented. We see that slope dramatically reduces at 7. From looking at slopes on the scree plot we would likely want to retain 6 components.

```r
summary(pca)
```

```r
var <- pca$sdev^2
porportionexplained <- var / sum(var)
```

```r
plot(cumsum(porportionexplained), xlab = 'Principle Component', ylab = 'Cumulative Porportion of Variance Captured', main = 'Cumulative Proportion of Variance Captured By Principle Component')
```

Cumulative proporton at the bottom shows us how much of the variance is accouunted for as the amount of variables used increases. We can see that 1 variable can account for 45% of the variance. At 10 variables, 95% of varaince within the data set is accounted for. Furthermore, at 17 varaibles we can accoutnt for 99% variance.Thus we would be using 17 varaibles instead of ~30 if this PCA were to be applied to models

```r
set.seed(1234)
bc1 = bc[,-1]
split <- sample.split(bc1$diagnosis, SplitRatio = 0.7)
train_data <- subset(bc1, split == TRUE)
test_data <-subset(bc1, split == FALSE)
```

splitting data into training and testing sets 

```r
train_data
test_data
```

```r
library(caret)
```

```r
fit = train(diagnosis ~., method = 'rpart', data = train_data)
library(rpart)
rpart_fit <- rpart(diagnosis~., method = 'class', data = train_data)
library(rpart.plot)
rpart.plot(rpart_fit)
```

In the cell above, I am using rpart to fit a cart model, and furthermore disaply a decision tree from the cart model. The deciscion rules are displayed as yes(left split) or no (right split). In the tree above we see that our cart model is using the variables radius worst, cocave points worst, and smoothness worst to predict if the tumor is cancerous(m) or not(b). We'll now test the predictive power of this model on the training and testing sets.

```r
train.pred <- predict(fit, newdata=train_data)

tablea <- table(train.pred,train_data$diagnosis)
tablea
test.pred <- predict(fit, newdata=test_data)
tableb <- table(test.pred,test_data$diagnosis)
tableb
```

predicting training and testing data and creating confusion tables. 

```r



print(paste("accuracy = ",round((sum(diag(tablea))/sum(tablea))*100,2)))
print(paste("accuracy = ",round((sum(diag(tableb))/sum(tableb))*100,2)))
```

prediction power of this model is 91.81 accuracy on the testing data. Which isn't bad! I'll now test my luck with a random forest model. Essentially, it creates multiple decision trees (you can set the number of trees to be used with ntrees = #) and uses the combination of these trees to reach the classification outcomes 

```r
library(randomForest)
```

```r
set.seed(500)
a=c()
i=5
for (i in 3:8) {
  model3 <- randomForest(diagnosis ~ ., data = train_data, mtry = i, ntree =500, importance = TRUE)
  prediag <- predict(model3, test_data, type = "class")
  a[i-2] = mean(prediag == test_data$diagnosis)
}
a
plot(3:8,a)
```

The above code runs random forest at varying levels of mtry (3:8). The plot in the output shows how accuracy changes with differing levels of mtry. Mtry defines the number of variables sampled at each split. We see that our acuracy increases at 5, so I will use this as my model mtry value.


```r
set.seed(500)
rf_fit <- randomForest(diagnosis ~ ., method = 'class',mtry =5, data = train_data,importance = TRUE)
rf_fit
```

```r
varImpPlot(rf_fit)
```

for the random forest model we created, it appears that concave.points_worst is the most important variable in terms of mean decrease accuracy as well as mean decresae gini. Our five most important variables for this model, again both in terms of mean decrease accuracy and mean decrease gini are concave_points worst, perimeter worst, area worst, radius worst, and concave_points mean. Note that radius worst and concave points worst made up 2 of the 3 variables used in our Cart model. Must be important!

```r
 importance(rf_fit)
```

Getting a better look at the importance values for these varaibles in our random forest model

```r
set.seed(500)
RF_fit <- train(diagnosis~ .,method="rf",data=train_data,tuneGrid = expand.grid(.mtry = 5))
```

training the random forest model to test its prediction accuracy on the training and testing sets
